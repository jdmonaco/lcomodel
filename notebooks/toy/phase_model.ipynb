## Distance nonlinearity
This function `g(d)` is a nonlinearity (hyperbolic tangent) of distance `d` from a preferred location. 
It is parameterized by a radius parameter `g_radius` that determines the smoothness and spatial scale of the approach.
The closer `g_radius` is to the size of the arena (80 cm) the more gradual the spatial activation.

```python
def g(d, g_radius=20, sigmoid=True):
    """ Compute a linear (or sigmoid) nonlinearity of distance """
    if sigmoid:
        return np.tanh(2 * (1 - d / g_radius))
    return 2 * (clip(g_radius - d / 2, 0, g_radius) / g_radius - 0.5)

# Plot g(d) for different values of g_radius
x = np.linspace(0,120,100) * cm
for r in array((10,15,20,30,45,60)) * cm:
    plt.plot(x / cm, g(x, g_radius=r), label='%d cm' % (r / cm))
plt.xlabel('distance (cm)')
plt.ylabel('g(distance)')
plt.legend();
```

## Model parameters
Population level parameters and values:

```python
N = 16
R_arena = 40.0 * cm
g_radius = 25 * cm
```

## LCO Neuron group
These are the equations and population of LCO neurons.

```python
eqs = """
    Phi = arccos(Ip) : radian
    Ip = tanh(2 * (1 - D / g_radius)) : 1
    D = sqrt((X - Xp)**2 + (Y - Yp)**2) : cm
    Xp : cm
    Yp : cm
    X : cm
    Y : cm
"""
LCO = NeuronGroup(N, model=eqs)

# Set the preferred locations for the LCO neurons
LCO.Xp = 2 * R_arena * (rand(N) - 0.5)
LCO.Yp = 2 * R_arena * (rand(N) - 0.5)
```

## Trajectory inputs and monitors
Create TimedArrays for the `(x,y)` trajectory data channels and then set up monitors for the model variables.

```python
# Add the timed trajectory data into the LCO population
LCO.X = TimedArray(x_pos, t_pos)
LCO.Y = TimedArray(y_pos, t_pos)
start_t = t_pos[0] * second

# Create the state monitors
Phis = StateMonitor(LCO, 'Phi', record=True)
Ips = StateMonitor(LCO, 'Ip', record=True)
Ds = StateMonitor(LCO, 'D', record=True)
Xs = StateMonitor(LCO, 'X', record=True)
Ys = StateMonitor(LCO, 'Y', record=True)
```

## Run the simulation and plot results

Update the default simulation clock to concur with the timing of the trajectory segment that we're using.

```python
reinit_default_clock()
defaultclock.dt = 10 * ms
defaultclock.start = start_t

run(120 * second)
```

## Plotting distance, spatial input, and phase

```python
f, ax = subplots(3, 1, sharex=True, figsize=(8,5))
ax1, ax2, ax3 = ax

# Plot distance from preferred location
ax1.plot(Ds.times, Ds[0] / cm)
ax1.set_ylabel('distance (cm)')
ax1.set_ylim(bottom=0)

# Plot activation (Ip) from Place/BVC/hippocampal inputs
ax2.plot(Ips.times, Ips[0], 'r-')
ax2.set_ylabel('activation')

# Plot phase angle of firing
ax3.plot(Phis.times, Phis[0], 'g-')
ax3.set_ylabel('phase (rad)')
ax3.set_xlabel('time (s)')
ax3.set_ylim(0, pi);
```

### Now, plot spatial phase of the LCO neurons in space.

```python
from tools.images import tiling_dims
r, c = tiling_dims(N)

f, ax = subplots(r, c, sharey=True, sharex=True, figsize=(12,12))
ax = ax.flatten()

# Phase-coded spatial trajectory plot
def plot_phase_trajectory(i, ax):
    ax.plot(LCO.Xp[i] / cm, LCO.Yp[i] / cm, 'k+', ms=20, mew=3)
    ax.scatter(Xs[0] / cm, Ys[0] / cm, c=Phis[i], cmap='hsv', vmin=0, vmax=pi, linewidths=0)
    ax.add_artist(mpl.patches.Circle((0,0), radius=R_arena / cm, fill=False, 
                                     ec='0.5', lw=2, ls='dashed'))
    ax.axis(array([-1,1,-1,1]) * 1.2 * R_arena / cm);

for i in xrange(N):
    plot_phase_trajectory(i, ax[i])
```
