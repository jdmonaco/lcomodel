```python
from brian import *
```

```python
tau = 20 * ms
N = 100
b = 1.2 # constant current mean
f = 10 * Hz
```

```python
print f
```

```python
eqs = '''
dv/dt=(-v+a*sin(2*pi*f*t)+b)/tau : 1
a : 1
'''
```

```python
neurons = NeuronGroup(N, model=eqs, threshold=1, reset=0)
neurons.v = rand(N)
neurons.a = linspace(.05, 0.75, N)
S = SpikeMonitor(neurons)
trace = StateMonitor(neurons, 'v', record=50)
```

```python
run(1000 * ms)
```

```python
raster_plot(S)
```

```python
#plt.figure(); 
plot(trace.times / ms, trace[50])
```
