## Boundary vector activity
This function `g(d)` is a nonlinearity (hyperbolic tangent) of distance `d` from a focal point, which will be used to emulate boundary-vector-cell activity. The focal point will act as a spatial phase or offset for a ring of activity about the same size as the environment. The smoothness of the sigmoid and the focus point (offset) are the parameters that vary between BVCs.

```python
def bvc(d, smoothness=20 * cm, D_arena=80 * cm, min_margin=0.1):
    return 0.5 + tanh(2 * (d - (1 - min_margin) * D_arena + smoothness) / smoothness) / 2

# Plot bvc(d) for different values of g_radius
x = linspace(0,100,100) * cm
for s in array((5,10,15,20,25,30)) * cm:
    plt.plot(x / cm, bvc(x, smoothness=s), label='%d cm' % (s / cm))
plt.xlabel('distance (cm)')
plt.ylabel('bvc(distance)')
plt.legend(loc='upper left');
```

## Model parameters
Population level parameters and values:

```python
R_arena = 40.0 * cm
D_arena = 2 * R_arena

N_BVC = 36
BVC_sigma = (10 * cm, 30 * cm)
BVC_min = 0.05
```

## Helper functions

This function sets an axis for the arena and draws the cylinder edge.

```python
def set_arena(ax=None):
    if ax is None:
        ax = gca()
    ax.add_artist(mpl.patches.Circle((0,0), radius=R_arena / cm, fill=False, ec='0.3', 
                                     lw=2, ls='dashed', zorder=-1))
    ax.axis(array([-1,1,-1,1]) * 1.2 * R_arena / cm)
    return ax
```

This function uniformly samples a circular disc for spatial phase offsets.

```python
def randc(N, radius):
    t = 2 * pi * rand(N)
    r = rand(N) + rand(N)
    r[r > 1] = 2 - r[r > 1]
    r *= radius
    return np.c_[r * cos(t), r * sin(t)].T

sample = randc(1000, 0.9 * R_arena) / cm
scatter(*sample, marker='.', color='b')
axis('equal')
set_arena();
```

## BVC neuron group
These are the equations for the BVC groups of neurons:

```python
bvc_eqs = """
    f = 0.5 + tanh(2 * (D - (1 - BVC_min) * D_arena + sigma) / sigma) / 2 : 1
    D = sqrt((X - Xp)**2 + (Y - Yp)**2) : cm
    sigma : cm
    Xp : cm # x-dimension spatial offset
    Yp : cm # y-dimension spatial offset
    X : cm
    Y : cm
"""
BVC = NeuronGroup(N_BVC, model=bvc_eqs)

BVC.Xp, BVC.Yp = randc(N_BVC, 0.9 * R_arena)
BVC.sigma = BVC_sigma[0] + diff(BVC_sigma) * rand(N_BVC)
```

Plot the BVC parameters.

```python
from tools.plot import quicktitle
f, ax = subplots(1, 2, figsize=(9,3))
ax1, ax2 = ax

ax1.scatter(BVC.Xp / cm, BVC.Yp / cm, marker='.', color='b')
set_arena(ax1)
ax1.axis([-70,70,-50,50])
quicktitle(ax1, 'spatial offsets')

ax2.hist(BVC.sigma / cm)
ax2.set_xlabel('sigma (cm)')
quicktitle(ax2, 'border spread');
```

## Trajectory inputs and monitors
Create TimedArrays for the `(x,y)` trajectory data channels and then set up monitors for the model variables.

```python
# Add the timed trajectory data into the LCO population
BVC.X = TimedArray(x_pos, t_pos)
BVC.Y = TimedArray(y_pos, t_pos)
start_t = t_pos[0] * second

# Create the state monitors
f_ = StateMonitor(BVC, 'f', record=True)
D_ = StateMonitor(BVC, 'D', record=True)
X_ = StateMonitor(BVC, 'X', record=True)
Y_ = StateMonitor(BVC, 'Y', record=True)
```

## Reset the clock and run the simulation

Update the default simulation clock to concur with the timing of the trajectory segment that we're using.

```python
reinit_default_clock()
defaultclock.dt = 10 * ms
defaultclock.start = start_t

run(180 * second)
```

## Plotting distance and BVC firing rate

```python
f, ax = subplots(2, 1, sharex=True, figsize=(8,5))
ax1, ax2 = ax

# Plot distance from preferred location
ax1.plot(D_.times, D_[0] / cm)
ax1.set_ylabel('distance (cm)')
ax1.set_ylim(bottom=0)

# Plot firing rate
ax2.plot(f_.times, f_[0], 'r-')
ax2.set_ylabel('firing rate (au)');
```

### Now, plot the firing rate profile of the BVC neurons in space.

```python
from tools.images import tiling_dims
r, c = tiling_dims(N_BVC)

f, ax = subplots(r, c, sharey=True, sharex=True, figsize=(12,12))
ax = ax.flatten()

# Phase-coded spatial trajectory plot
def plot_phase_trajectory(i, ax):
    ax.plot(BVC.Xp[i] / cm, BVC.Yp[i] / cm, 'm+', ms=20, mew=3)
    ax.scatter(X_[0] / cm, Y_[0] / cm, c=f_[i], cmap='jet', linewidths=0)
    ax.axis('equal')
    set_arena(ax)

for i in xrange(N_BVC):
    plot_phase_trajectory(i, ax[i])
```
