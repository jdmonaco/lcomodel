## BVC neuron group

Set up the BVC neuron equations and group:

```python
N_BVC = 36
BVC_sigma = 15 * cm
BVC_sigma_min = 15 * cm
BVC_margin = 0.1

bvc_eqs = """
    f = 0.5 + tanh(2 * (D - (1 - BVC_margin) * D_arena + sigma) / sigma) / 2 : 1
    D = sqrt((X - Xp)**2 + (Y - Yp)**2) : meter
    sigma : meter
    Xp : meter
    Yp : meter
    X : meter (linked)
    Y : meter (linked)
"""

BVC = NeuronGroup(N_BVC, model=bvc_eqs)

BVC.X = linked_var(Trajectory, 'X')
BVC.Y = linked_var(Trajectory, 'Y')

BVC.Xp, BVC.Yp = arena_sample(N_BVC)
BVC.sigma = BVC_sigma_min + BVC_sigma * rand(N_BVC)
```

Show BVC parameter units and distributions:

```python
def plot_bvc_parameters():
    f, ax = subplots(1,2, figsize=(13,4))
    ax1, ax2 = ax
    f.suptitle('BVC neuron parameters')

    ax1.scatter(BVC.Xp / cm, BVC.Yp / cm, marker='.', color='r')
    draw_arena(ax1)
    ax1.axis('equal')
    ax1.axis([-70,70,-50,50])
    ax1.set_xlabel('reference locations')
    print('unit(Xp) =', get_unit(BVC.Xp))

    ax2.hist(BVC.sigma / cm)
    ax2.set_xlabel('smoothness (sigma)')
    print('unit(sigma) =', get_unit(BVC.sigma))

plot_bvc_parameters()
```

## Boundary vector activity
This function `bvc(d)` is a nonlinearity (hyperbolic tangent) of distance `d` from a focal point, which will be used to emulate boundary-vector-cell activity. The focal point will act as a spatial phase or offset for a ring of activity about the same size as the environment. The smoothness of the sigmoid and the focus point (offset) are the parameters that vary between BVCs.

```python
def bvc(D, sigma=20 * cm):
    return 0.5 + tanh(2 * (D - (1 - BVC_margin) * D_arena + sigma) / sigma) / 2

# Plot bvc(d) for different values of g_radius
def plot_bvc_function():
    x = linspace(0,100,100) * cm
    for s in np.linspace(BVC_sigma_min, BVC_sigma_min + BVC_sigma, 8):
        plt.plot(x / cm, bvc(x, sigma=s), label='%d cm' % (s / cm))
    plt.xlabel('distance (cm)')
    plt.ylabel('bvc(distance)')
    plt.legend(loc='upper left');
plot_bvc_function()
```

### Now, plot the firing rate profile of the BVC neurons in space.

```python
r, c = tiling_dims(N_BVC)

f, ax = subplots(r, c, sharey=True, sharex=True, figsize=(12,12))
ax = ax.flatten()

# Phase-coded spatial trajectory plot
def plot_phase_trajectory(i, ax):
    ax.plot(BVC.Xp[i] / cm, BVC.Yp[i] / cm, 'm+', ms=20, mew=3)
    ax.scatter(posmon.X / cm, posmon.Y / cm, c=mon.f[i], cmap='jet', linewidths=0)
    ax.axis('equal')
    draw_arena(ax)

for i in xrange(N_BVC):
    plot_phase_trajectory(i, ax[i])
```
