##Load the trajectory data

```python
import os
datadir = '/Users/joe/source/phase_model/data'
posfile = os.path.join(datadir, 'pos.npy')
pos = np.load(posfile)
x_pos, y_pos = pos[1:3] * cm
t_pos = (pos[0] - pos[0][0]) * second
```

## Model parameters
Population level parameters and values:

```python
R_arena = 40.0 * cm
D_arena = 2 * R_arena

N_BVC = 48
BVC_sigma = (15 * cm, 25 * cm)
BVC_min = 0.05

N_LCO = 6
chi = 0.6 * mA
C_in = 4
```

## Helper functions

This function sets an axis for the arena and draws the cylinder edge.

```python
def set_arena(ax=None):
    if ax is None:
        ax = gca()
    ax.add_artist(mpl.patches.Circle((0,0), radius=R_arena / cm, fill=False, ec='0.3', 
                                     lw=2, ls='dashed', zorder=-1))
    ax.axis(array([-1,1,-1,1]) * 1.2 * R_arena / cm)
    return ax
```

This function uniformly samples a circular disc for spatial phase offsets.

```python
def randc(N, radius):
    t = 2 * pi * rand(N)
    r = rand(N) + rand(N)
    r[r > 1] = 2 - r[r > 1]
    r *= radius
    return np.c_[r * cos(t), r * sin(t)].T
```

## Neuron groups

These are the equations for the BVC group of neurons:

```python
bvc_eqs = """
    f = 0.5 + tanh(2 * (D - (1 - BVC_min) * D_arena + sigma) / sigma) / 2 : 1
    D = sqrt((X - Xp)**2 + (Y - Yp)**2) : cm
    sigma : cm
    Xp : cm # x-dimension spatial offset
    Yp : cm # y-dimension spatial offset
    X : cm
    Y : cm
"""
BVC = NeuronGroup(N_BVC, model=bvc_eqs)

BVC.Xp, BVC.Yp = randc(N_BVC, R_arena)
BVC.sigma = BVC_sigma[0] + diff(BVC_sigma) * rand(N_BVC)
```

```python
W = zeros((N_BVC, N_LCO), 'd')
for j in xrange(N_LCO):
    con = permutation(N_BVC)[:C_in]
    W[con,j] = 1.0
print('C_in = ', W.sum(axis=0), '(should be %d)' % C_in)
```

## Trajectory inputs and monitors

```python
# Add the timed trajectory data into the BVC population
BVC.X = TimedArray(x_pos, t_pos)
BVC.Y = TimedArray(y_pos, t_pos)
start_t = t_pos[0] * second

# Create the state monitors
f_ = StateMonitor(BVC, 'f', record=True)
D_ = StateMonitor(BVC, 'D', record=True)
X_ = StateMonitor(BVC, 'X', record=True)
Y_ = StateMonitor(BVC, 'Y', record=True)
```

### Reset the clock and run the simulation

```python
reinit_default_clock()
defaultclock.dt = 100 * ms
defaultclock.start = start_t

run(300 * second)
```

### Post-hoc computation of LCO firing rate and phase

```python
I = dot(f_[:].T, W).T 
V = tanh(2 * (I - (chi / mA)) / (chi / mA))
f_LCO = clip(V, 0, 1)
Theta = arccos(V)
```

### Plot cell-pairs phase-difference maps

```python
f, ax = subplots(N_LCO, N_LCO, sharey=True, sharex=True, figsize=(15,15))

def plot_phase_trajectory(cdata, ax, cmap='jet', vmin=0, vmax=pi):
    ax.scatter(X_[0] / cm, Y_[0] / cm, c=cdata, cmap=cmap, vmin=vmin, vmax=vmax, linewidths=0)
    ax.axis('equal')
    set_arena(ax)

for i in xrange(N_LCO):
    for j in xrange(N_LCO):
        if j < i:
            ax[i][j].set_axis_off()
        elif j == i:
            plot_phase_trajectory(Theta[i], ax[i][j], cmap='copper_r')
            quicktitle(ax[i][j], 'LCO %d' % i)
        else:
            plot_phase_trajectory(cdiff(Theta[i], Theta[j]), ax[i][j], cmap='hsv', vmin=-pi)
            quicktitle(ax[i][j], '%d x %d' % (i, j))
```
